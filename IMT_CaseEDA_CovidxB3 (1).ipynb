## Gráficos
### Análise dos casos de COVID-19

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

Analizando os dados verificamos que existem valores negativos

```python
df[df.regiao == 'Brasil'].casosNovos.nsmallest(5)
```

Então para uma análise mais precisa foram realizados os seguintes filtros:
- Somente os dados considerando o Brasil como um todo (ignorando separação de regiões)

```python
# Gráfico dos casos de COVID-19
df[(df.regiao == 'Brasil') & (df.casosNovos >= 0)].plot(x="data", y="casosNovos", figsize=(12, 6))
plt.xlabel("Data")
plt.ylabel("Casos novos")
plt.title("Casos novos de COVID-19 no Brasil")
plt.grid(True)

# Linha de Tendência
x = np.arange(len(df[(df.regiao == 'Brasil') & (df.casosNovos >= 0)]["data"]))
y = df[(df.regiao == 'Brasil') & (df.casosNovos >= 0)]["casosNovos"]
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")

plt.show()
```

### Análise das ações

Faremos a análise do valor de fechamento das ações (Adj Close)

```python
# Gráfico das ações
df_fleury.plot(y="Adj Close", figsize=(12, 6))
plt.xlabel("Data")
plt.ylabel("Valor da Ação (USD)")
plt.title("Valor da ação do Fleury")
plt.grid(True)
```

```python
# Linha de tendência
x = np.arange(len(df_fleury))
y = df_fleury["Adj Close"]
plt.figure(figsize=(12, 6))
plt.xlabel("Data")
plt.ylabel("Valor da Ação (USD)")
plt.title("Valor da ação do Fleury")
plt.grid(True)
# Mostrar Datas
stride = 50  # Passo em que as datas são mostradas
plt.xticks(x[::stride], df_fleury.index[::stride], rotation=45)

# Linha de tendência polinomial
z1 = np.polyfit(x, y, 1)
p1 = np.poly1d(z1)
# Linha de tendência de 2° grau
z2 = np.polyfit(x, y, 2)
p2 = np.poly1d(z2)
plt.plot(x,p1(x),"r--")
plt.plot(x,p2(x),"y--")
plt.plot(x,y)

plt.show()
```

```python
# Grafico das Ações
df_azul.plot(y="Adj Close", figsize=(12, 6))
plt.xlabel("Data")
plt.ylabel("Valor da Ação (USD)")
plt.title("Valor da ação da Azul")
plt.show()
```

```python
# Linha de tendência
x = np.arange(len(df_azul))
y = df_azul["Adj Close"]
plt.figure(figsize=(12, 6))
plt.xlabel("Data")
plt.ylabel("Valor da Ação (USD)")
plt.title("Valor da ação da Azul")
plt.grid(True)
# Mostrar Datas
stride = 50  # Passo em que as datas são mostradas
plt.xticks(x[::stride], df_azul.index[::stride], rotation=45)

# Linha de tendência polinomial
z1 = np.polyfit(x, y, 1)
p1 = np.poly1d(z1)
# Linha de tendência de 2° grau
z2 = np.polyfit(x, y, 2)
p2 = np.poly1d(z2)
plt.plot(x,p1(x),"r--")
plt.plot(x,p2(x),"y--")
plt.plot(x,y)

plt.show()
```

### Comparação dos valores


```python
# Criação da coluna "data" para facilitar a junção dos dataframes
df_fleury['data']= df_fleury.index
df_fleury.head()
```

```python
# Criação da coluna "data" para facilitar a junção dos dataframes
df_azul['data']= df_azul.index
df_azul.head()
```

```python
df['data'] = pd.to_datetime(df['data'])
```

```python
# Gráficos juntados F.O.

df_merged = pd.merge(df_fleury, df_azul, on='data')
df_merged = pd.merge(df_merged ,df[(df.regiao == 'Brasil') & (df.casosNovos >= 0)], how='right', on='data') #Adicionado o método de merge
df_merged['Adj Close_x'].fillna(method='ffill', inplace=True, axis=0) #Adicionada para preencher com dados do dia anterior
df_merged['Adj Close_y'].fillna(method='ffill', inplace=True, axis=0) #Adicionada para preencher com dados do dia anterior
#df_merged.head()

fig, ax1 = plt.subplots(figsize=(12, 6))

# Plotar casos novos de COVID-19 no eixo principal
ax1.plot(df_merged['data'], df_merged['casosNovos'], color='b', label='Casos Novos')
ax1.set_xlabel('Data')
ax1.set_ylabel('Casos Novos', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Plotar valor das ações no eixo secundário
ax2 = ax1.twinx()
ax2.plot(df_merged['data'], df_merged['Adj Close_x'], color='g', label='Fleury')
ax2.plot(df_merged['data'], df_merged['Adj Close_y'], color='r', label='Azul')
ax2.set_ylabel('Valor da Ação (USD)', color='g')
ax2.tick_params(axis='y', labelcolor='g')

# Adicionar título e legenda
plt.title('Comparação de casos novos de COVID-19 com o valor de ações do Fleury e da Azul')
plt.legend()

plt.show()
```
